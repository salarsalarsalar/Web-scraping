```python
import heapq

def dijkstra(graph, start):
    # Create a priority queue to store vertices and their distances
    priority_queue = [(0, start)]
    
    # Initialize distances with infinity for all vertices except the start vertex
    distances = {vertex: float('infinity') for vertex in graph}
    distances[start] = 0
    
    while priority_queue:
        # Get the vertex with the smallest distance
        current_distance, current_vertex = heapq.heappop(priority_queue)
        
        # Check if the current distance is already greater than the known distance
        if current_distance > distances[current_vertex]:
            continue
        
        # Explore neighbors of the current vertex
        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight
            
            # If a shorter path is found, update the distance and add to the priority queue
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))
    
    return distances

# Example usage
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}

start_vertex = 'A'
shortest_distances = dijkstra(graph, start_vertex)

print(f"Shortest distances from {start_vertex}:")
for vertex, distance in shortest_distances.items():
    print(f"To {vertex}: {distance}")
```

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt

def dijkstra(graph, start):
    priority_queue = [(0, start)]
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0
    
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        
        if current_distance > distances[current_node]:
            continue
        
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))
    
    return distances

def visualize_network(graph, shortest_path):
    G = nx.Graph(graph)
    
    pos = nx.spring_layout(G)  # You can choose different layout algorithms based on your preference
    
    nx.draw(G, pos, with_labels=True, font_weight='bold')
    
    # Highlight the nodes and edges in the shortest path
    edge_labels = {}
    for i in range(len(shortest_path) - 1):
        edge_labels[(shortest_path[i], shortest_path[i + 1])] = graph[shortest_path[i]][shortest_path[i + 1]]
    
    nx.draw_networkx_nodes(G, pos, nodelist=shortest_path, node_color='r')
    nx.draw_networkx_edges(G, pos, edgelist=edge_labels.keys(), edge_color='r', width=2)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    
    plt.show()

# Example network topology represented as a weighted graph
network_topology = {
    'Network_A': {'Network_B': 1, 'Network_C': 4},
    'Network_B': {'Network_A': 1, 'Network_C': 2, 'Network_D': 5},
    'Network_C': {'Network_A': 4, 'Network_B': 2, 'Network_D': 1},
    'Network_D': {'Network_B': 5, 'Network_C': 1}
}

# Specify source and destination networks
source_network = 'Network_A'
destination_network = 'Network_D'

# Run Dijkstra's Algorithm
shortest_distances = dijkstra(network_topology, source_network)

# Retrieve shortest path to the destination network
shortest_path = []
current_network = destination_network
while current_network != source_network:
    shortest_path.insert(0, current_network)
    current_network = min(network_topology[current_network], key=lambda x: shortest_distances[x] + network_topology[current_network][x])

shortest_path.insert(0, source_network)

# Display the result
print(f"Shortest path from {source_network} to {destination_network}: {shortest_path}")
print(f"Total distance: {shortest_distances[destination_network]}")

# Visualize the network and the shortest path
visualize_network(network_topology, shortest_path)
```

```python
import heapq
import tkinter as tk
import time
import random

class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = {}
        
    def add_node(self, value):
        self.nodes.add(value)
        self.edges[value] = {}
        
    def add_edge(self, from_node, to_node, weight):
        self.edges[from_node][to_node] = weight
        self.edges[to_node][from_node] = weight  # Assuming an undirected graph

def dijkstra(graph, start):
    priority_queue = [(0, start)]
    distances = {node: float('infinity') for node in graph.nodes}
    distances[start] = 0
    
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        
        if current_distance > distances[current_node]:
            continue
        
        for neighbor, weight in graph.edges[current_node].items():
            distance = current_distance + weight
            
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))
    
    return distances

def visualize_graph(graph):
    root = tk.Tk()
    root.title("Graph Visualization")

    canvas = tk.Canvas(root, width=400, height=400)
    canvas.pack()

    # Assign positions to nodes
    node_positions = {node: (50 * i + 15, 165) for i, node in enumerate(graph.nodes)}

    # Draw nodes
    for node, (x, y) in node_positions.items():
        canvas.create_oval(x - 15, y - 15, x + 15, y + 15, fill="blue")
        canvas.create_text(x, y, text=node, fill="white")

    # Draw edges
    for from_node in graph.edges:
        for to_node, weight in graph.edges[from_node].items():
            x1, y1 = node_positions[from_node]
            x2, y2 = node_positions[to_node]
            canvas.create_line(x1, y1, x2, y2, width=2, arrow=tk.LAST, fill="black")

    root.mainloop()

def generate_random_graph(num_nodes):
    graph = Graph()
    for i in range(num_nodes):
        graph.add_node(f'Node_{i}')
    
    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):
            if random.choice([True, False]):  # Randomly decide if there is a connection
                weight = random.randint(1, 10)  # Random weight for the connection
                graph.add_edge(f'Node_{i}', f'Node_{j}', weight)
    
    return graph

# Measure and display the performance for different graph sizes
for num_nodes in [10, 20, 30, 40, 50]:
    # Generate a random graph with the specified number of nodes
    graph = generate_random_graph(num_nodes)
    
    # Choose random source node
    source_node = random.choice(list(graph.nodes))

    # Run Dijkstra's Algorithm and measure the execution time
    start_time = time.time()
    shortest_distances = dijkstra(graph, source_node)
    end_time = time.time()
    execution_time = end_time - start_time

    # Display the result and performance metrics
    print(f"Graph size: {num_nodes}")
    print(f"Execution time: {execution_time:.6f} seconds\n")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import heapq

def dijkstra(graph, start):
    # Initialize distances and predecessors
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0
    predecessors = {node: None for node in graph}
    
    # Priority queue to store nodes with their distances
    priority_queue = [(0, start)]
    
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        
        # Check if the current distance is smaller than the stored distance
        if current_distance > distances[current_node]:
            continue
        
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                predecessors[neighbor] = current_node
                heapq.heappush(priority_queue, (distance, neighbor))
    
    return distances, predecessors

def visualize(graph, distances, predecessors):
    G = nx.Graph(graph)  # Convert the dictionary to a NetworkX graph
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue')
    
    for node, cost in distances.items():
        plt.annotate(f'{node}\n{cost}', xy=pos[node], xytext=(10, -20), textcoords='offset points', ha='center', fontsize=8)
    
    for node, predecessor in predecessors.items():
        if predecessor is not None:
            plt.arrow(pos[predecessor][0], pos[predecessor][1], pos[node][0]-pos[predecessor][0], pos[node][1]-pos[predecessor][1], 
                      length_includes_head=True, head_width=0.02, color='red')
    
    plt.show()

# Example graph
graph = {
    'A': {'B': 1, 'D': 3},
    'B': {'A': 1, 'C': 2, 'D': 4},
    'C': {'B': 2, 'D': 5},
    'D': {'A': 3, 'B': 4, 'C': 5}
}

# Starting node
start_node = 'A'

# Run Dijkstra's algorithm
distances, predecessors = dijkstra(graph, start_node)

# Visualize the graph with distances and predecessors
visualize(graph, distances, predecessors)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import heapq

def dijkstra(graph, start):
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0
    predecessors = {node: None for node in graph}
    
    priority_queue = [(0, start)]
    
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        
        if current_distance > distances[current_node]:
            continue
        
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                predecessors[neighbor] = current_node
                heapq.heappush(priority_queue, (distance, neighbor))
    
    return distances, predecessors

def visualize(graph, distances, predecessors):
    G = nx.Graph(graph)
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue')
    
    for node, cost in distances.items():
        plt.annotate(f'{node}\n{cost}', xy=pos[node], xytext=(10, -20), textcoords='offset points', ha='center', fontsize=8)
    
    for node, predecessor in predecessors.items():
        if predecessor is not None:
            plt.arrow(pos[predecessor][0], pos[predecessor][1], pos[node][0]-pos[predecessor][0], pos[node][1]-pos[predecessor][1], 
                      length_includes_head=True, head_width=0.02, color='red')
    
    plt.show()

def add_edge(graph, node1, node2, weight):
    if node1 not in graph:
        graph[node1] = {}
    if node2 not in graph:
        graph[node2] = {}
    graph[node1][node2] = weight
    graph[node2][node1] = weight

def remove_edge(graph, node1, node2):
    if node1 in graph and node2 in graph[node1]:
        del graph[node1][node2]
    if node2 in graph and node1 in graph[node2]:
        del graph[node2][node1]

def add_node(graph, node):
    if node not in graph:
        graph[node] = {}

def remove_node(graph, node):
    if node in graph:
        del graph[node]
        for other_node in graph:
            if node in graph[other_node]:
                del graph[other_node][node]

# Example graph
graph = {
    'A': {'B': 1, 'D': 3},
    'B': {'A': 1, 'C': 2, 'D': 4},
    'C': {'B': 2, 'D': 5},
    'D': {'A': 3, 'B': 4, 'C': 5}
}

# Starting node
start_node = 'A'

# Run Dijkstra's algorithm and visualize the initial graph
distances, predecessors = dijkstra(graph, start_node)
visualize(graph, distances, predecessors)

# Add a new node and edge
add_node(graph, 'E')
add_edge(graph, 'D', 'E', 2)

# Run Dijkstra's algorithm and visualize the updated graph
distances, predecessors = dijkstra(graph, start_node)
visualize(graph, distances, predecessors)

# Remove an edge
remove_edge(graph, 'B', 'C')

# Run Dijkstra's algorithm and visualize the updated graph
distances, predecessors = dijkstra(graph, start_node)
visualize(graph, distances, predecessors)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import heapq
import timeit
def generate_random_graph(size):
    G = nx.gnm_random_graph(size, size * 2)  # Generate a random graph with nodes=size and edges=size*2
    graph = {node: {neighbor: G[node][neighbor].get('weight', 1) for neighbor in G.neighbors(node)} for node in G.nodes}
    return graph

sizes = [10, 20, 30, 40, 50]  # You can adjust the sizes as needed

for size in sizes:
    graph = generate_random_graph(size)
    
    # Choose a random starting node
    start_node = list(graph.keys())[0]
    
    # Measure execution time using %timeit
    time_taken = timeit.timeit(lambda: dijkstra(graph, start_node), number=1)
    
    print(f"Graph size: {size}, Execution time: {time_taken:.6f} seconds")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import networkx as nx
import heapq

class DijkstraVisualizer:
    def __init__(self, master):
        self.master = master
        self.master.title("Dijkstra's Algorithm Visualizer")

        self.graph = {}
        self.start_node = tk.StringVar()
        self.start_node.set("")
        self.create_widgets()

    def create_widgets(self):
        # Label and Entry for entering the starting node
        ttk.Label(self.master, text="Starting Node:").grid(row=0, column=0, padx=10, pady=5)
        entry_start_node = ttk.Entry(self.master, textvariable=self.start_node)
        entry_start_node.grid(row=0, column=1, padx=10, pady=5)

        # Button to run Dijkstra's algorithm
        ttk.Button(self.master, text="Run Dijkstra", command=self.run_dijkstra).grid(row=0, column=2, padx=10, pady=5)

        # Matplotlib Figure for graph visualization
        self.figure, self.ax = plt.subplots(figsize=(6, 6))
        self.canvas = FigureCanvasTkAgg(self.figure, master=self.master)
        self.canvas_widget = self.canvas.get_tk_widget()
        self.canvas_widget.grid(row=1, column=0, columnspan=3, padx=10, pady=5)

        # Entry and Button for dynamically creating the graph
        ttk.Label(self.master, text="Graph Creation:").grid(row=2, column=0, padx=10, pady=5)
        self.graph_entry = ttk.Entry(self.master)
        self.graph_entry.grid(row=2, column=1, padx=10, pady=5)
        ttk.Button(self.master, text="Create Graph", command=self.create_graph).grid(row=2, column=2, padx=10, pady=5)

    def run_dijkstra(self):
        start_node = self.start_node.get()

        if not start_node:
            messagebox.showerror("Error", "Please enter a starting node.")
            return

        if start_node not in self.graph:
            messagebox.showerror("Error", f"Starting node '{start_node}' not found in the graph.")
            return

        distances, predecessors = dijkstra(self.graph, start_node)
        self.visualize_graph(distances, predecessors)

    def visualize_graph(self, distances=None, predecessors=None):
        self.ax.clear()
        G = nx.Graph(self.graph)
        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', ax=self.ax)

        if distances and predecessors:
            for node, cost in distances.items():
                self.ax.annotate(f'{node}\n{cost}', xy=pos[node], xytext=(10, -20), textcoords='offset points',
                                 ha='center', fontsize=8)

            for node, predecessor in predecessors.items():
                if predecessor is not None:
                    self.ax.arrow(pos[predecessor][0], pos[predecessor][1], pos[node][0] - pos[predecessor][0],
                                  pos[node][1] - pos[predecessor][1],
                                  length_includes_head=True, head_width=0.02, color='red')

        self.canvas.draw()

    def create_graph(self):
        try:
            graph_str = self.graph_entry.get()
            self.graph = eval(graph_str)  # Use eval to convert the string to a dictionary

            if not isinstance(self.graph, dict):
                raise ValueError("Invalid graph format. Please use a valid dictionary format.")

            # Visualize the initial graph
            self.visualize_graph()

        except Exception as e:
            messagebox.showerror("Error", f"Error creating graph: {e}")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import networkx as nx
import heapq
import random
import string
import time

def dijkstra(graph, start):
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0
    pq = [(0, start)]
    
    while pq:
        current_distance, current_node = heapq.heappop(pq)
        
        if current_distance > distances[current_node]:
            continue
        
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
    
    predecessors = {node: None for node in graph}
    for node in graph:
        if distances[node] == float('infinity'):
            continue
        current_node = node
        while predecessors[current_node] is not None:
            current_node = predecessors[current_node]
        predecessors[node] = current_node

    return distances, predecessors

class DijkstraVisualizer:
    def __init__(self, master):
        self.master = master
        self.master.title("Dijkstra's Algorithm Visualizer")

        self.graph = {}
        self.start_node = tk.StringVar()
        self.start_node.set("A")

        self.performance_results = []

        self.create_widgets()

    def create_widgets(self):
        # Label and Entry for entering the starting node
        ttk.Label(self.master, text="Starting Node:").grid(row=0, column=0, padx=10, pady=5)
        entry_start_node = ttk.Entry(self.master, textvariable=self.start_node)
        entry_start_node.grid(row=0, column=1, padx=10, pady=5)

        # Button to run Dijkstra's algorithm
        ttk.Button(self.master, text="Run Dijkstra", command=self.run_dijkstra).grid(row=0, column=2, padx=10, pady=5)

        # Matplotlib Figure for graph visualization
        self.figure, self.ax = Figure(figsize=(6, 6)), None
        self.canvas = FigureCanvasTkAgg(self.figure, master=self.master)
        self.canvas_widget = self.canvas.get_tk_widget()
        self.canvas_widget.grid(row=1, column=0, columnspan=3, padx=10, pady=5)

        # Button for adding a random edge to the graph
        ttk.Button(self.master, text="Add Random Edge", command=self.add_random_edge).grid(row=2, column=0, padx=10, pady=5)

        # Button for removing a random edge from the graph
        ttk.Button(self.master, text="Remove Random Edge", command=self.remove_random_edge).grid(row=2, column=1, padx=10, pady=5)

        # Button for generating a random graph
        ttk.Button(self.master, text="Generate Random Graph", command=self.generate_random_graph).grid(row=2, column=2, padx=10, pady=5)

        # Button for measuring performance
        ttk.Button(self.master, text="Measure Performance", command=self.measure_performance).grid(row=2, column=3, padx=10, pady=5)

        # Initial visualization
        self.visualize_graph()

    def run_dijkstra(self):
        start_node = self.start_node.get()

        if start_node not in self.graph:
            messagebox.showerror("Error", f"Starting node '{start_node}' not found in the graph.")
            return

        distances, predecessors = dijkstra(self.graph, start_node)
        self.visualize_graph(distances, predecessors)

    def visualize_graph(self, distances=None, predecessors=None):
            if self.ax is not None:
                self.ax.clear()
            else:
                self.ax = self.figure.add_subplot(111)

            G = nx.Graph(self.graph)
            pos = nx.spring_layout(G)
            nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', ax=self.ax)

            if distances and predecessors:
                edge_labels = {(node, neighbor): weight for node, neighbors in self.graph.items() for neighbor, weight in neighbors.items()}
                nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, ax=self.ax, font_color='red')

                for node, cost in distances.items():
                    self.ax.annotate(f'{node}\n{cost}', xy=pos[node], xytext=(10, -20), textcoords='offset points',
                                     ha='center', fontsize=8)

                for node, predecessor in predecessors.items():
                    if predecessor is not None:
                        self.ax.arrow(pos[predecessor][0], pos[predecessor][1], pos[node][0] - pos[predecessor][0],
                                      pos[node][1] - pos[predecessor][1],
                                      length_includes_head=True, head_width=0.02, color='red')

            self.canvas.draw()


    def add_random_edge(self):
        nodes = list(self.graph.keys())
        node1, node2 = random.sample(nodes, 2)
        weight = random.randint(1, 10)

        if node2 not in self.graph[node1]:
            self.graph[node1][node2] = weight
            self.graph[node2][node1] = weight
            self.visualize_graph()

    def remove_random_edge(self):
        nodes = list(self.graph.keys())
        node1, node2 = random.sample(nodes, 2)

        if node2 in self.graph[node1]:
            del self.graph[node1][node2]
            del self.graph[node2][node1]
            self.visualize_graph()

    def generate_random_graph(self):
        num_nodes = random.randint(5, 15)
        num_edges = random.randint(num_nodes, num_nodes * 2)

        alphabet_names = list(string.ascii_uppercase)[:num_nodes]

        G = nx.gnm_random_graph(num_nodes, num_edges)
        self.graph = {alphabet_names[node]: {alphabet_names[neighbor]: random.randint(1, 10) for neighbor in G.neighbors(node)} for node in G.nodes}

        self.visualize_graph()

    def measure_performance(self):
        # Measure performance for different network sizes
        network_sizes = [5, 10, 15, 20]
        self.performance_results = []

        for size in network_sizes:
            # Generate a random graph
            self.generate_random_graph()

            # Randomly select a starting node
            start_node = random.choice(list(self.graph.keys()))

            # Measure execution time for Dijkstra's algorithm
            start_time = time.time()
            dijkstra(self.graph, start_node)
            end_time = time.time()

            # Record the performance results
            self.performance_results.append((size, end_time - start_time))

        # Display the performance results
        self.visualize_graph()  # Add this line to display the graph after measuring performance
        self.display_performance_results()

    def display_performance_results(self):
        if not self.performance_results:
            messagebox.showinfo("Information", "Run Dijkstra's algorithm first to measure performance.")
            return

        result_text = "Performance Results:\n\n"
        for size, execution_time in self.performance_results:
            result_text += f"Network Size: {size}, Execution Time: {execution_time:.6f} seconds\n"

        messagebox.showinfo("Performance Results", result_text)

if __name__ == "__main__":
    root = tk.Tk()
    app = DijkstraVisualizer(root)
    root.mainloop()
```
