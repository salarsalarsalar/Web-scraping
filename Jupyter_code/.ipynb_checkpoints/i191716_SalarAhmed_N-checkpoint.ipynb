## Question 2

In this example we will be using polymorphic with Hello world as our example
Techniques for encyption and decryption discussed below

```python
import random

# Original string
original_text = "Hello, World!"

# Generate a random key
key = random.randint(1, 255)

# Function to polymorphically encrypt the string
def polymorphic_encrypt(text, key):
    encrypted_text = ""
    for char in text:
        encrypted_char = ord(char) ^ key  # XOR operation with the key
        encrypted_text += chr(encrypted_char)
    return encrypted_text

# Function to polymorphically decrypt the string
def polymorphic_decrypt(encrypted_text, key):
    decrypted_text = ""
    for char in encrypted_text:
        decrypted_char = ord(char) ^ key  # XOR operation with the same key
        decrypted_text += chr(decrypted_char)
    return decrypted_text

# Encrypt the string
encrypted_hello = polymorphic_encrypt(original_text, key)
print("Encrypted:", encrypted_hello)

# Decrypt the string
decrypted_hello = polymorphic_decrypt(encrypted_hello, key)
print("Decrypted:", decrypted_hello)
```

```python
def caesar_cipher(text, shift):
    transformed_text = ""
    for char in text:
        if char.isalpha():
            # Determine if it's an uppercase or lowercase letter
            is_upper = char.isupper()
            
            # Shift the letter by the specified number of positions
            char_code = ord(char)
            shifted_code = char_code + shift
            
            # Ensure the letter wraps around the alphabet
            if is_upper:
                if shifted_code > ord('Z'):
                    shifted_code -= 26
                elif shifted_code < ord('A'):
                    shifted_code += 26
            else:
                if shifted_code > ord('z'):
                    shifted_code -= 26
                elif shifted_code < ord('a'):
                    shifted_code += 26
            
            transformed_text += chr(shifted_code)
        else:
            # Non-alphabet characters remain unchanged
            transformed_text += char
    
    return transformed_text

original_text = "Hello, World!"
shift = 3

transformed_text = caesar_cipher(original_text, shift)
print("Original Text:", original_text)
print("Transformed Text:", transformed_text)
```
