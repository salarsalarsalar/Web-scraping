```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

URL = "https://www.eduvision.edu.pk/programs-offered-in-engineering-at-bachelor-level-in-pakistan"

# Initialize WebDriver and open the URL
driver = webdriver.Chrome()
driver.get(URL)


# Locate the dropdown element
dropdown = driver.find_element(By.XPATH, "//select[@class='form-control' and @name='disciplineType']")
university_links = driver.find_elements(By.XPATH ,f'//*[@id="main"]/div/div/div[1]/div[3]/div/div[2]/ul/li/div[2]/a[1]')


# Create an empty dictionary to store ranks for each option
rank_data = {}

# Iterate over each option and click on it
select = Select(dropdown)


for index in range(1, len(select.options)):
    option = select.options[index]
    
    # Click on the option
    option.click()
    
    # Extract the ranks for the current option
    ul_element = driver.find_element(By.XPATH, "//ul[contains(@class, 'sc_list') and contains(@class, 'sc_list_style_star')]")
    university_ranks = [li.text for li in ul_element.find_elements(By.TAG_NAME, "li")]
    
    
    # Store the ranks in the dictionary
    rank_data[option.text] = university_ranks
```
